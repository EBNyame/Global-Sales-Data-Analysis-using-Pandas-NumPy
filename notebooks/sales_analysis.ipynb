The `Postal Code` column has missing value because some countries do not use postal codes.The missing values are not random but systematic and it is very important.

Since it is not random and it's also importand, we can't drop it or fill the unkown with mean or median so we will just set the unkown values to `Unkown`

```python
df['Postal Code'] = df['Postal Code'].fillna("Unkown")
df
```
