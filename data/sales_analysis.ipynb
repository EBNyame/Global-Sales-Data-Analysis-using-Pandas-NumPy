The `Postal Code` column has missing value because some countries do not use postal codes.The missing values are not random but systematic and it is very important.

Since it is not random and it's also importand, we can't drop it or fill the unkown with mean or median so we will just set the unkown values to `Unkown`

```python
df['Postal Code'] = df['Postal Code'].fillna("Unkown")
df
```

**Converting date columns to proper datetime format**

I will convert the date if the datee is invalid to `NaT` (Not a Time). This will prevents the notebook from crashing. I can extract year/month/day which makes time-series analysis becomes easy.

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")
```

```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"], errors="coerce")
```

```python
# checking if it worked
df[["Order Date", "Ship Date"]].head()
```

```python
df.info()
```

```python
df["Order Date"].isnull().sum()
```

```python
df["Ship Date"].isnull().sum()
```

**Extract date**

```python
# Extract the year
df["Order Year"] = df["Order Date"].dt.year
```

```python
# extract the month
df["Order Month"] = df["Order Date"].dt.month_name()
```

```python
# extract the day
df["Order Day"] = df["Order Date"].dt.day_name()
```

```python
# checking the dataset
df
```

```python
df.columns
```

**Calculate shipping time (in days):**

```python
df["Shipping Time(Days)"] = (df['Ship Date'] - df['Order Date']).dt.days
```

## Date Handling

The `Order Date` and `Ship Date` columns were converted from string

features such as order year, month, and shipping duration were
